# Image Pyramids

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

from custom_utils import *

import logging

logging.basicConfig(
    #  filename='log_file_name.log',
     level=logging.INFO, 
     format= '[%(asctime)s] %(levelname)s - %(message)s',
     datefmt='%H:%M:%S'
 )
```

```python
height, width, img, img_rgb, img_gray = load_image_and_show("../data/wavesheep1.jpg")
```

```python
def recursive_pyramid(img, level):
    if level == 0:
        return img
    img = cv.pyrDown(img)
    return recursive_pyramid(img, level - 1)

level = 3
_, ax = plt.subplots(1, level, figsize=(width*level/100, height/100), sharey=True, sharex=True)
for i in range(3):
    title = f"PyrDown x{i}"
    show_image(recursive_pyramid(img_rgb, i), title, ax=ax[i])
    ax[i].set_ylim(height, 0)
    ax[i].set_xlim(width, 0)

```
