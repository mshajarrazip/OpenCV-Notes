# Canny Edge Detection

Edge detection algorithm!!!

[Canny Edge Detection](https://docs.opencv.org/4.x/da/d22/tutorial_py_canny.html)

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

from custom_utils import *

import logging
logging.basicConfig(
    #  filename='log_file_name.log',
     level=logging.INFO, 
     format= '[%(asctime)s] %(levelname)s - %(message)s',
     datefmt='%H:%M:%S'
 )
height, width, img, img_rgb, img_gray = load_image_and_show("../data/phobs1.jpeg")
```

```python
img_adaptive = cv.adaptiveThreshold(
    img_gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2
)
edges = cv.Canny(img_adaptive, height, width)
_, edges_flipped = cv.threshold(edges, 127, 255, cv.THRESH_BINARY_INV)

titles = ['Original', 'Adaptive Gaussian', 'Canny', 'Canny - Flipped']
_, ax = plt.subplots(1, 4, figsize=(width*4/100, height/100))
for i, im in enumerate([img_gray, img_adaptive, edges, edges_flipped]):
    show_image(im, ax=ax[i], title=titles[i], imshow_args={'cmap': 'gray'})
plt.show()
```
